```python
import pydicom
import matplotlib.pyplot as plt
import numpy as np

# Read the DICOM file
dcm_file1 = pydicom.read_file("./Patient 1 Quad View/_H163510/M5AGHIGS")
dcm_file2 = pydicom.read_file("./Patient 1 Quad View/_H163518/M5AGHKGU")

# Get the image data from the DICOM file
image_data1 = dcm_file1.pixel_array
image_data2 = dcm_file2.pixel_array

def to_gray(img):
    return img[...,:3] @ [0.33, 0.33, 0.33]

# Display the image
plt.imshow(to_gray(image_data1), cmap='gray')
plt.show()
plt.imshow(to_gray(image_data2), cmap='gray')
```

```python
import pydicom
import matplotlib.pyplot as plt

# List of file paths for the DICOM images
file_paths = ["./Patient 1 Quad View/_H163510/M5AGHIGS", 
              "./Patient 1 Quad View/_H163518/M5AGHKGU",
              "./Patient 1 Quad View/_H163526/M5AGHMH0",
              "./Patient 1 Quad View/_H163531/M5AGHNP2"]

# Number of rows and columns in the subplot grid
n_rows = 2
n_cols = 2

# Create the subplot grid
fig, axs = plt.subplots(n_rows, n_cols, figsize=(10, 10))

# Loop over the file paths
for i, file_path in enumerate(file_paths):
    # Read the DICOM file
    dcm_file = pydicom.read_file(file_path)

    # Get the image data from the DICOM file
    image_data = dcm_file.pixel_array

    # Calculate the subplot index
    row = i // n_cols
    col = i % n_cols

    # Plot the image data in the subplot
    axs[row, col].imshow(to_gray(image_data), cmap='gray')

# Display the figure
plt.show()
```
